# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = '../data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
app_train['OWN_CAR_AGE'].describe()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
print (app_train.dtypes.value_counts())
##dtype_select = ([np.dtype['int'], np.dtype['float'])
dtype_select = ['float64', 'int64']
numeric_columns= list(app_train.select_dtypes(include=dtype_select))                 
print ("numeric_columns=", numeric_columns, "length=", len(numeric_columns))

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns %s" % len(numeric_columns))
app_train["OBS_60_CNT_SOCIAL_CIRCLE"]= ~pd.isna(app_train["OBS_60_CNT_SOCIAL_CIRCLE"])
i=0
# 檢視這些欄位的數值範圍
for col in numeric_columns:
    """
    Your CODE HERE, make the box plot
    """
    ##print("col=", col, "Nan count=", pd.isna(app_train[col]).value_counts('True'))
    nan_count=pd.isna(app_train[col]).sum()
    if (nan_count < 1):
        #print(app_train[col].describe())
        #print(np.percentile(app_train[col], 25))
        #rint(np.percentile(app_train[col], 75))
        print("column=", col, "Nan data count=", nan_count )
        print(app_train[col].describe())
        plt.boxplot(app_train[col])
        plt.show()
        i=i+1
        print ("field number w/o nan=", i)
    else:
        print ("column=", col, "Nan data count=", nan_count)
        print (app_train[col].describe())
        filtered_data= ~pd.isna(app_train[col])
        plt.boxplot(filtered_data)      
        plt.show()
        i=i+1
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
sigma=app_train["AMT_INCOME_TOTAL"].std()
mu = 200
n_bins = 50
filtered_data= app_train[app_train["AMT_INCOME_TOTAL"] <= sigma*6]
#print (filtered_data.value_counts())
print ("filtered_data", filtered_data["AMT_INCOME_TOTAL"].describe())
print ("index=", filtered_data.index)
plt.xlim(0, len(app_train.index))
plt.scatter(filtered_data.index, filtered_data["AMT_INCOME_TOTAL"])
#from statsmodels.distributions.empirical_distribution import ECDF
#cdf = ECDF (app_train["AMT_INCOME_TOTAL"])
#hist, bin_edges=np.histogram(app_train["AMT_INCOME_TOTAL"], bins=30 , normed=True)
n, bins, patches = plt.hist(filtered_data["AMT_INCOME_TOTAL"], n_bins, density=True, histtype='step', cumulative=True, label='Empirical')
#plt.hist(filtered_data["AMT_INCOME_TOTAL"], density=True)
print ("n=", n, "bin=", bin, "patches=", patches)
plt.show()
cdf=pd.DataFrame(n)
#y=((1/(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*(1/sigma*(bins-mu))**2))
#print ("Y_type=", type(y), "Y=", y)
#y = y.cumsum()
#cdf=pd.DataFrame(y)

print ("cdf.index=",cdf.index )
#cdf = y/ y[-1] 

#plt.plot(cdf)
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
