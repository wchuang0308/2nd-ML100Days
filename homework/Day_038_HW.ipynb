```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, roc_auc_score, f1_score
```

```python
wine = datasets.load_wine()
wine_data = wine.data
#print (wine.DESCR)
feature_names=['Alcohol', 'Malic_acid', 'Ash', 'Alcalinity_of_ash','Magnesium', 'Total_phenols', 'Flavanoids', 'Nonflavanoid_phenols', 'Proanthocyanins', 'Color_intensity', 'Hue', 'OD280_OD315_of_diluted_wines', 'Proline']
target = wine.target
target = target.astype(np.float64)
print ("data=", wine_data, "target=", target)
wine_df_data = pd.DataFrame(wine_data, columns=feature_names)
target_df = pd.DataFrame({'Target':target})
wine_df_data.loc[:, 'Target'] = target
wine_df_data.describe()
```

```python
#print (wine_df_data['Target'])
wine_df_data.corr()
```

```python

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.1, random_state=4)

# 建立模型
logreg = linear_model.LogisticRegression(solver='lbfgs', max_iter=10000)

# 訓練模型
logreg.fit(x_train, y_train)

# 預測測試集
y_pred = logreg.predict(x_test)
y_pred_proba = logreg.predict(x_test)
accuracy = accuracy_score( y_test, y_pred)
print("y_pred", y_pred, 'y_test=', y_test, 'accuracy=', accuracy)
```

```python
boston = datasets.load_boston()
boston_data = boston.data
#print (boston.DESCR)
feature_names=['CRIM', 'ZN', 'INDUS', 'CHAS','NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
price = boston.target
#target = t.astype(np.float64)
print ("data=", boston_data[0:4], "price=", price[0:4])
boston_df_data = pd.DataFrame(boston_data, columns=feature_names)
boston_df_data.loc[:, 'PRICE'] = price
boston_df_data.describe()
boston_df_data.corr()
```

```python
# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(boston_data, price, test_size=0.3, random_state=4)

# 建立一個線性回歸模型
regr = linear_model.LinearRegression()

# 將訓練資料丟進去模型訓練
regr.fit(x_train, y_train)

# 將測試資料丟進模型得到預測結果
y_pred = regr.predict(x_test)
MSE = mean_squared_error(y_test, y_pred)
print ("RMSE=", MSE**0.5)
```

```python
breast_cancer = datasets.load_breast_cancer()
breast_cancer_data = breast_cancer.data
print (breast_cancer.DESCR)
feature_names=['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean','smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave_points_mean', 'symmetry_mean', 'fractal_dimension_mean','radius_stde', 'texture_stde', 'perimeter_stde', 'area_stde','smoothness_stde','compactness_stde', 'concavity_stde', 'concave_points_stde', 'symmetry_stde', 'fractal_dimension_stde', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst','smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave_points_worst', 'symmetry_worst', 'fractal_dimension_worst']
target = breast_cancer.target
#target = t.astype(np.float64)
print ("data=", boston_cancer_data[0:4], "classfication=", target[0:4])
breast_cancer_df_data = pd.DataFrame(breast_cancer_data, columns=feature_names)
breast_cancer_df_data.loc[:, 'Target'] = target
breast_cancer_df_data.describe()
breast_cancer_df_data.corr()
```

```python
# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, test_size=0.3, random_state=4)

# 建立模型
logreg = linear_model.LogisticRegression(solver='lbfgs', max_iter=10000)

# 訓練模型
logreg.fit(x_train, y_train)

# 預測測試集
y_pred = logreg.predict(x_test)
y_pred_proba = logreg.predict(x_test)
accuracy = accuracy_score( y_test, y_pred)
print("y_pred", y_pred, 'y_test=', y_test, 'accuracy=', accuracy)
```

```python
y_pred = logreg.predict(x_test)
auc = roc_auc_score(y_test, y_pred) # 使用 roc_auc_score 來評估。 **這邊特別注意 y_pred 必須要放機率值進去!**
print("AUC: ", auc) 
```
